```python
from scipy.stats import zscore
import seaborn as sns
import numpy as np
data=sns.load_dataset('iris')
df=data.copy()
```

```python
df
```

```python
df['sepal_length'].mean()/df['sepal_length'].std()
```

```python
z_score = np.abs(zscore(df.drop('species',axis=1)))
z_score
```

```python
zscore(df.drop('species',axis=1))
```

```python
non_outliers = (z_score < 3).all(axis=1)
df_non_outliers = df[non_outliers]
```

```python
df_non_outliers
```

```python
outliers = (z_score > 3).any(axis=1)
outliers_rows =df[outliers]
outliers_rows
```
